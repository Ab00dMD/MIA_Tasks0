# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Load the Dataset

```python
data = pd.read_csv('shrink_ray_dataset.csv')
```

# Data Preprocessing

```python
power = data[['Power']]
shrinkage = data[['Shrinkage']]
```

# Create and Train the Linear Regression Model

```python
model = LinearRegression()
model.fit(power, shrinkage)
```

# Where should he adjust the knob?

```python
knob_required = (85 - model.intercept_) / model.coef_
print(knob_required)
```

# BONUS

```python
# Create a scatter plot of the dataset.
plt.scatter(power, shrinkage, label='Data', color='red')
# Plot the linear model.
plt.plot(power, model.predict(power), label='Linear Model', color='black')
# Add labels
plt.xlabel('Power (MW)')
plt.ylabel('Percentage (%)')
# Add a legend
plt.legend()
# Show the grid
plt.grid(True)
# Show the plot
plt.show()
```
